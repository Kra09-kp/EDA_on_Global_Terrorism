## Data Cleaning and Exploration

```python
t_data = data[['eventid','iyear', 'imonth', 'iday', 'country_txt','city','provstate','region_txt','attacktype1_txt','weaptype1_txt','targtype1_txt','gname','targsubtype1_txt','success','nkill']]
```

```python
t_data.rename(columns={'eventid':"Id",'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','city':'City','provstate':'State','region_txt':'Region','attacktype1_txt':'Attacktype','weaptype1_txt':'Weapontype','targtype1_txt':'Targettype','gname':'Group','targsubtype1_txt':'Targetsubtype','success':'Success','nkill':'Killed'},inplace=True)
t_data.head()
```

```python
t_data.describe()
```

```python
t_data.shape
```

```python
t_data.info()
```

```python
t_data.isnull().sum().sort_values(ascending=False)
```

```python
t_data.dtypes
```

```python
t_data['Targetsubtype'].fillna("Others",inplace=True)
t_data['City'].fillna("Unknown",inplace=True)
t_data['State'].fillna("Unknown",inplace=True)
t_data['Killed'].fillna(0,inplace=True)
t_data.head()
```

```python
t_data.isnull().sum()
```

```python
t_data.drop_duplicates(inplace=True)
```

```python
# Number of total countries
total_countries = t_data['Country'].unique()
print('Their are total {} countries where terror attack happened.'.format(len(total_countries)))
pd.DataFrame(total_countries).rename(columns = {0:'Country'})
```

```python
# Number of total region
t_region = t_data['Region'].unique()
print('Their are total {} region where terror attack happened.'.format(len(t_region)))
pd.DataFrame(t_region).rename(columns = {0:'Region'})
```

```python
t_data.drop('Id',axis=1,inplace=True)
t_data.head()
```

```python
t_data.shape
```

```python
#Most attacked Countries
most_att_countries = t_data['Country'].value_counts().sort_values(ascending=False).head(15)
print('Most attacked country is " {} " with the number of " {} " .'.format(most_att_countries.idxmax(),most_att_countries.max()))
plt.figure(figsize=(15,10))
plt.xticks(rotation=45)
plt.plot(most_att_countries)
```

```python
#Region,in which most attack happened 
most_att_region = t_data['Region'].value_counts().sort_values(ascending=False).head(5)
print('Most of the attacks are happen in the " {} " with the number of " {} " .'.format(most_att_region.idxmax(),most_att_region.max()))
plt.figure(figsize=(15,10))
plt.xticks(rotation=45)
plt.plot(most_att_region)
```

```python
#Groups,by which most attacks done and top 5 attack groups 
most_att_group = t_data['Group'].value_counts().sort_values(ascending=False).head(5)
print('"{}" did most of the attacks.'.format(most_att_group.index[1]))
most_att_group
```

```python
#Top 15 states where most of the terrorist activities occur
terror_state =  t_data['State'].value_counts().sort_values(ascending=False).head(15)
print(terror_state)
plt.figure(figsize=(15,10))
plt.xticks(rotation=45)
plt.plot(terror_state)
```

```python
#top 15 cities which are most affected by terrorist activities
terror_cities =  t_data['City'].value_counts().sort_values(ascending=False).head(15)
print(terror_cities)
```

```python
#top 6 attack type which are occur most
most_att_type = t_data['Attacktype'].value_counts().sort_values(ascending=False).head(7)
most_att_type
```

```python
#top 15 target type which are targeted most of the time
most_tar_type = t_data['Targettype'].value_counts().sort_values(ascending=False).head(15)
most_tar_type
```

```python
#top 15 sub target type which are targeted most of the time
most_subtar_type = t_data['Targetsubtype'].value_counts().sort_values(ascending=False).head(15)
most_subtar_type
```

## Data Visualisation

```python
plt.figure(figsize=(15,10))
plt.title("Terrorist activities with respect to year")
plt.xticks(rotation = 60)
sns.countplot(t_data.Year)
```

```python
plt.figure(figsize=(15,10))
plt.title("Terrorist activities with respect to Region and Success")
plt.xticks(rotation = 60)
sns.countplot(t_data.Region,hue = t_data.Success)
```

```python
plt.figure(figsize=(15,10))
plt.title("Terrorist activities with respect to month")
sns.countplot(t_data.Month)
```

```python
plt.figure(figsize=(15,10))
plt.title("Terrorist activities with respect to Attact type")
plt.xticks(rotation = 60)
sns.countplot(t_data.Attacktype)
```

```python
t_data.Weapontype.replace({'Vehicle (not to include vehicle-borne explosives, i.e., car or truck bombs)':'Vehicles'},inplace=True)
```

```python
plt.figure(figsize=(15,10))
plt.title("Terrorist activities with respect to Weapon type")
plt.xticks(rotation = 60)
sns.countplot(t_data.Weapontype)
```

```python
plt.figure(figsize=(13,9))
plt.title("Terrorist activities with respect to Targettype type")
sns.countplot(y=t_data.Targettype)
```

```python
pd.crosstab(t_data.Year,t_data.Region).plot(kind='area',figsize=(20,10))
```

```python
sns.heatmap(np.round(t_data.corr(),4),annot=True,cmap='coolwarm')
plt.yticks(rotation=90)
```
